## 2. Dataset

```python
DATA_YAML = "../dataset/data.yaml"
```

## 3. Treinamento

Escolha a variante do YOLOv11 (`n`, `s`, `m`, `l`) e os hiperparâmetros. O treino salva os pesos e as métricas em `runs/detect/<name>`.

```python
from ultralytics import YOLO

# Variante: yolov11n.pt (nano), yolov11s.pt (small), yolov11m.pt (medium), yolov11l.pt (large)
MODEL_VARIANT = "yolov11n.pt"
EPOCHS = 50
IMGSZ = 640
BATCH = 16
RUN_NAME = "yolov11n-urban-disaster"

model = YOLO(MODEL_VARIANT)
model.train(
    data=DATA_YAML,
    epochs=EPOCHS,
    imgsz=IMGSZ,
    batch=BATCH,
    name=RUN_NAME,
)
```

```python
# Pesos treinados ficam em runs/detect/<RUN_NAME>/weights/best.pt
# Para exportar: model.export(format="onnx")  # ou format="engine" para TensorRT
```

## 4. Validar em uma imagem

Após o treino, teste o modelo em uma imagem de validação.

```python
from pathlib import Path

weights_path = Path("runs/detect") / RUN_NAME / "weights" / "best.pt"

if weights_path.exists():
    model = YOLO(str(weights_path))
    
    # Predição em uma imagem (substitua pelo caminho de uma imagem do dataset)
    results = model.predict(source="../dataset/valid/images/", save=True)
    
    print(f"Modelo carregado: {weights_path}")
else:
    print("Execute o treinamento acima primeiro.")
```
