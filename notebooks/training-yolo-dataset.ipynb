## 2. Dataset

```python
DATA_YAML = "../dataset/data.yaml"
```

## 3. Training

Choose the YOLOv11 variant (`n`, `s`, `m`, `l`) and hyperparameters. Training saves weights and metrics under `runs/detect/<name>`.

```python
from ultralytics import YOLO

# Variant: yolov11n.pt (nano), yolov11s.pt (small), yolov11m.pt (medium), yolov11l.pt (large)
MODEL_VARIANT = "yolov11n.pt"
EPOCHS = 50
IMGSZ = 640
BATCH = 16
RUN_NAME = "yolov11n-urban-disaster"

model = YOLO(MODEL_VARIANT)
model.train(
    data=DATA_YAML,
    epochs=EPOCHS,
    imgsz=IMGSZ,
    batch=BATCH,
    name=RUN_NAME,
)
```

```python
# Trained weights are saved under runs/detect/<RUN_NAME>/weights/best.pt
# To export: model.export(format="onnx")  # or format="engine" for TensorRT
```

## 4. Validate on images

After training, test the model on validation images.

```python
from pathlib import Path

weights_path = Path("runs/detect") / RUN_NAME / "weights" / "best.pt"

if weights_path.exists():
    model = YOLO(str(weights_path))
    
    # Predict on images (replace with a path to a dataset image or folder)
    results = model.predict(source="../dataset/valid/images/", save=True)
    
    print(f"Model loaded: {weights_path}")
else:
    print("Run the training cell above first.")
```
